```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y=load_iris(return_X_y=True)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.columns
```

```python
df.shape
```

```python
for column in df.columns:
    print(df[column].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
# Initialize LabelEncoder
sex_label_encoder = LabelEncoder()
smoker_label_encoder = LabelEncoder()
```

```python
# Apply Label Encoding
df['sex'] = sex_label_encoder.fit_transform(df['sex'])
```

```python
# Apply Label Encoding
df['smoker'] = smoker_label_encoder.fit_transform(df['smoker'])
```

```python
print(df['sex'].unique())
print(df['smoker'].unique())
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import pandas as pd
```

```python
# Initialize OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse_output=False)

# Apply One-Hot Encoding
encoded_data = one_hot_encoder.fit_transform(df[['day']])

# Convert to DataFrame with proper column names
df_one_hot = pd.DataFrame(encoded_data, columns=one_hot_encoder.get_feature_names_out(['day']))

# Concatenate the new columns to the original DataFrame
df = pd.concat([df, df_one_hot], axis=1)

# Drop the original column
df = df.drop(columns=['day'])
```

```python
df.head(2)
```

```python
X = df[["total_bill", "tip", "sex", "smoker", "size", "day_Fri", "day_Sat", "day_Sun", "day_Thur"]]
y = df["time"]
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
